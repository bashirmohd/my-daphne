```python
import json
from datetime import datetime
```

```python
def get_data(pathway, startDate, endDate):
    data = {}
    for date in range(startDate, endDate):
        print(date)
        fltr = RowKeyRegexFilter(r'1d-%s::esnet\/prod::devices\/%s\/.+?::traffic\/in' % (str(date), pathway))
        for key, row in table.scan(filter = fltr):
            timestamp_dict = row.keys() #get all timestamps from data
            for timestamp in timestamp_dict:
                timestamp_cleaned = timestamp[3:] #[3:] to eliminate the "ts:" at beginning of each key
                if timestamp_cleaned in data: 
                    try:
                        data[timestamp_cleaned] += float(row[timestamp])
                    except:
                        continue
                else:
                    try:
                        data[timestamp_cleaned] = float(row[timestamp])
                    except:
                        continue
        with open(pathway+'.json', 'a') as f:
            f.write(json.dumps(data, ensure_ascii=False))
        f.close()
        data = {}
                        
    #return data
```

```python
def get_all_site_data(date, direction):
    data = {}
    fltr = RowKeyRegexFilter(r'1d-%s::esnet\/prod::devices\/.+?\/.+?::traffic\/%s' % (str(date), direction))
    for key, row in table.scan(filter = fltr):
        
        #get interface name
        match = re.search(r'1d-17700::esnet\/prod::devices\/.+?\/', key)
        site = match.group(0)
        site = site[30:]
        site = site[:-1]
        
        #check if site in data
        if site in data:
            timestamp_dict = row.keys() #get all timestamps from data
            for timestamp in timestamp_dict:
                timestamp_cleaned = timestamp[3:] #[3:] to eliminate the "ts:" at beginning of each key
                if timestamp_cleaned in data: 
                    try:
                        data[site][timestamp_cleaned] += float(row[timestamp])
                    except:
                        continue
                else:
                    try:
                        data[site][timestamp_cleaned] = float(row[timestamp])
                    except:
                        continue
        else:
            data[site] = {}
            timestamp_dict = row.keys() 
            for timestamp in timestamp_dict:
                timestamp_cleaned = timestamp[3:] 
                if timestamp_cleaned in data: 
                    try:
                        data[site][timestamp_cleaned] += float(row[timestamp])
                    except:
                        continue
                else:
                    try:
                        data[site][timestamp_cleaned] = float(row[timestamp])
                    except:
                        continue
    return data
```
