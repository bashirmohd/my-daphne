```python

corr_list = []
t = 24 * 30
leng = int(df.shape[0]/t)
pca = PCA(n_components=15)
for i in range(leng):
    start = i * t
    end = (i+1) * t
    df_temp = df.iloc[start:end]
    df_corr = df_temp.corr().round(2).abs()
    df_corr = df_corr.fillna(0).values 
    if i == 0:
        df_random1000 = df.sample(n = 1000) 
        pca.fit(df_random1000)
        df_pca = pca.transform(df_corr).flatten()
    else:
        df_pca = pca.transform(df_corr).flatten()
    corr_list.append(df_pca)
corr_mat = np.array(corr_list)
print(np.sum(pca.explained_variance_))
print(np.sum(pca.explained_variance_ratio_))
print(corr_mat.shape)

tsne = TSNE(n_components=2, random_state=0)
X_2d_month = tsne.fit_transform(corr_mat)
```

```python
plt.rcParams.update({'font.size':15})
plt.figure(figsize=(12, 8))
t = np.arange(X_2d_month.shape[0])
plt.scatter(X_2d_month[:,0], X_2d_month[:,1], s=140, c=t , cmap='RdBu')
plt.colorbar()
for i in range(X_2d_month.shape[0]):
    plt.annotate(i+1, (X_2d_month[i,0]+3, X_2d_month[i,1] + 3), fontsize=15)
plt.title('Correlation 2D monthly plot')
```
