```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from pylab import *
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 40})
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
rc('axes', linewidth=4)
```

```python
df = pd.read_csv('tcpboxplot_1024MB.csv')
```

```python
df.head(5)
```

```python
fig = plt.gcf()

# Changing Seaborn Plot size
fig.set_size_inches(21, 13)

# Setting the font scale
sns.set(style="ticks",font_scale=3.5, rc={"lines.linewidth": 3})
sns.boxplot(x="Data Size", y=" Aggregated Throughput(Gbps)", hue="Data Streams",data=df, palette="colorblind")
rc('axes', linewidth=4)
plt.title('S = 1024MB')
plt.show()

```
