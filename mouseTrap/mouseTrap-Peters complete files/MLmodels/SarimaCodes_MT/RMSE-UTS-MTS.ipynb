```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

matplotlib.rcParams.update({'font.size': 22})

plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams['xtick.labelsize']=20
plt.rcParams['ytick.labelsize']=20
```

```python
path_to_data = 'https://raw.githubusercontent.com/bashirmohd/DataSceince/master/Time-series-prediction_production/testlinks/ta_rmse.csv'

```

```python
df = pd.read_csv(path_to_data)
```

```python
df = df.set_index('Links')

```

```python
df = df.dropna()
```

```python
df.plot.bar(figsize=(12,8))
plt.xlabel('Network Links')
plt.ylabel('Prediction Error')
#plt.title('Bandwidth Utitilization Over Time (3-Months)')
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams.update({'font.size': 2})
plt.show()
```
