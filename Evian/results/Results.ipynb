```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


N = 2
#Times in mimutes
handlingTicket = (12, 0.2)
understandRequest = (0.5, 0.5)
installIntent = (12, 0.5)
ind = np.arange(N)    
widthBar = 0.35       

p1 = plt.bar(ind, handlingTicket, edgecolor='white', width=widthBar, color='#d62728')
p2 = plt.bar(ind, understandRequest, edgecolor='white', width=widthBar,
             bottom=handlingTicket)
p3 = plt.bar(ind, installIntent, edgecolor='white', width=widthBar,
             bottom=[i+j for i,j in zip(understandRequest,handlingTicket)])

plt.ylabel('Time in hours')
plt.title('Times to fulfill a transfer intent')
plt.xticks(ind, ('Without Evian', 'With Evian'))
plt.legend((p1[0], p2[0], p3[0]), ('Time before the ticket is handled', 'Time to understand the request', 
                                   'Time to install the intent'))

plt.show()
plt.savefig('times.png')
```
